```python
features = pd.DataFrame(index=df.index)
```

```python
def convert_date(input_string):
    """changes date from format 'Mon-YYYY' to
    'MM-YYYY'
    """
    months_num = {
        'Jan' : 1,
        'Feb' : 2,
        'Mar' : 3,
        'Apr' : 4,
        'May' : 5,
        'Jun' : 6,
        'Jul' : 7,
        'Aug' : 8,
        'Sep' : 9,
        'Oct' : 10,
        'Nov' : 11,
        'Dec' : 12
    }
    mon = input_string[:3]
    year = int(input_string[-4:])
    if mon in months_num:
        num = months_num[mon]
    else:
        raise ValueError('{} not found in dictionary'.format(mon))
    months_passed = (12 - num) + ((2015 - year) * 12)
    if not str(months_passed).isnumeric():
        raise Exception('Error: return object not numeric: {}'.format(months_passed))
    return months_passed

def term_numeric(input_string):
    """Takes loan term (string) and returns the term length
    in number of months (numeric)
    """
    return int(input_string[1:3])
```

```python
features['income_log'] = np.log(df.income) #these distributions definitely look more normal on log scale.
features['interest_paid_log'] = np.log(df.interest_paid)
features['principal_paid_log'] = np.log(df.principal_paid)
```

```python
sns.kdeplot(features.income_log);
```

```python
sns.kdeplot(features.interest_paid_log); #long tail on the left because of recently-issued loans
```

```python
sns.kdeplot(features.principal_paid_log); # same effect causing long tail
```

```python
features['months_since_issue'] = df.issue_date.apply(convert_date) #number of months since issue
```

```python
features['loan_term'] = df.loan_term.apply(term_numeric)
```

```python
features['term_ended'] = np.where((features.months_since_issue <= features.loan_term), 1, 0)
```

```python
features['outstanding_balance'] = df.loan_size - df.principal_paid
```

```python
features['profit'] = df.interest_paid - features.outstanding_balance
```

```python
features['avg_monthly_yield'] = features.profit / features.months_since_issue
```

```python
crisis_months = list(range(78, 97)) #months of sub-prime mortgage crisis
features['crisis_issued'] = np.where((features.months_since_issue.isin(crisis_months)), 1, 0) #1 if issued during crisis
```

```python
features['before_crisis'] = np.where((features.months_since_issue > 96), 1, 0) #1 if issued before crisis
```

```python
features['after_crisis'] = np.where((features.months_since_issue < 78), 1, 0) #1 if issued after crisis
```
